```python
import pandas as pd
import unicodedata
import os 

# Carregar Dados 
funcionarios = pd.read_csv('C:\\Users\\LUIZA\\analise_dados_tech_solutions\\data\\funcionarios.csv', encoding='ISO-8859-1', sep=',')
beneficios = pd.read_csv('C:\\Users\\LUIZA\\analise_dados_tech_solutions\\data\\beneficios.csv', encoding='ISO-8859-1', sep=',')

#verificar os nomes das colunas
print(funcionarios.columns)
print(beneficios.columns)
     
#Função para padronizar colunas
def normalizar_colunas(df):
    df.columns = (
        df.columns
        .str.strip()
        .map(lambda x: unicodedata.normalize('NFKD', x)
             .encode('ascii', errors='ignore')
             .decode('utf-8'))
        .str.replace(' ', '_')
        .str.replace('[()]', '', regex=True)
    )
    return df

# Aplicar função nos dois DataFrames
funcionarios = normalizar_colunas(funcionarios)
beneficios = normalizar_colunas(beneficios)

# Renomear colunas para nomes finais mais claros
funcionarios = funcionarios.rename(columns={
    'ID_do_FuncionArio': 'ID_do_Funcionario',
    'SalArio_r$': 'Salario',
    'Tempo_de_ServiAo_Prestado_anos': 'Tempo_de_Servico_anos',
    'SatisfaAAo_%': 'Satisfacao'
})

beneficios = beneficios.rename(columns={
    'Tipo_de_BenefAcio': 'Tipo_de_Beneficio',
    'DescriAAo': 'Descricao',
    'Taxa_de_AdesAo_%_do_SalArio': 'Taxa_de_Adesao',
    'Valor_Fixo_r$': 'Valor_Fixo'
})

print(funcionarios.columns)
print(beneficios.columns)

#Verificar os dados carregados 
print(funcionarios.head())
print(beneficios.head())

# Análise Descritiva
print(funcionarios.describe())
print(beneficios.describe())

import seaborn as sns
import matplotlib.pyplot as plt

# Definir estilo
sns.set_style("whitegrid") #ou "darkgrid", "ticks"
sns.set_context("talk") # aumenta o tamanho da fonte 

#Gráfico de Salário por Cargo 
plt.figure(figsize=(12, 6))
sns.barplot(
    x='cargo',
    y='Salario',
    data=funcionarios,
    palette='viridis' #cores bonitas
)
plt.title('Salários por Cargo', fontsize=16, weight='bold')
plt.xlabel('Cargo', fontsize=14)
plt.ylabel('Salário (R$)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  # rotaciona nomes para caber
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig(f'{directory}/salarios_por_cargo.png')  # SALVA A IMAGEM
plt.show()

# Gráfico de Satisfação 
plt.figure(figsize=(12, 6))
sns.histplot(
    funcionarios['Satisfacao'],
    bins=10,
    kde=True,
    color='skyblue'
    )

plt.title('Distribuição de Satisfação dos Funcionários', fontsize=16, weight='bold')
plt.xlabel('Satisfação(%)', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig(f'{directory}/satisfacao_funcionarios.png')
plt.show()
 
directory='../visualizations'
if not os.path.exists(directory):
    os.makedirs(directory)


```
